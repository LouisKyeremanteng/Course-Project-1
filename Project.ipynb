### Accessing each column separately
- Making a pie chat for each column

```python
fig = plt.figure(figsize=(10,14), dpi=160)
explode = [0,0.1]
lb= ['Below', 'Above']

# For fixed acidity
x= dataset['fixed acidity'] < dataset['fixed acidity'].mean()
ax1 = plt.subplot2grid((4,3),(0,0))
plt.pie(x.value_counts(), colors=("y","g"),labels= lb, autopct = '%.2f%%', explode= explode)
plt.title('fixed_acidity')

# For volatile acidity
y= dataset['volatile acidity'] < dataset['volatile acidity'].mean()
ax2 = plt.subplot2grid((4,3),(0, 1))
plt.pie(y.value_counts(),colors=("y","g"),labels= lb, autopct = '%.2f%%', explode= explode)
plt.title('volatile_acidity')

# For citric acid
z = dataset['citric acid'] < dataset['citric acid'].mean()
ax3 = plt.subplot2grid((4,3),(0, 2))
plt.pie(z.value_counts(),colors=("y","g"),labels= lb, autopct = '%.2f%%', explode= explode)
plt.title('citric_acid')

#For residual sugar
a= dataset['residual sugar'] < dataset['residual sugar'].mean()
ax1 = plt.subplot2grid((4,3),(1,0))
plt.pie(a.value_counts(),colors=("y","g"), autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('residual_sugar')

# For chlorides
c=dataset.chlorides<dataset.chlorides.mean()
ax1 = plt.subplot2grid((4,3), (1, 1))
plt.pie(c.value_counts() ,colors=("y","g"), autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('chlorides')

#For sulphur dioxide
d= dataset['free sulfur dioxide']<dataset['free sulfur dioxide'].mean()
ax1 = plt.subplot2grid((4,3),(1,2))
plt.pie(d.value_counts() ,colors=("y","g"), autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('free_sulfur_dioxide')

# For total sulphur dioxide
e= dataset['total sulfur dioxide']< dataset['total sulfur dioxide'].mean()
ax1 = plt.subplot2grid((4,3),(2, 0))
plt.pie(e.value_counts(),colors=("y","g"),autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('total_sulfur_dioxide')

#For density
f= dataset.density < dataset.density.mean()
ax1 = plt.subplot2grid((4,3),(2, 1))
plt.pie(f.value_counts(),colors=("y","g"), autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('density')

#For pH
g= dataset.pH<dataset.pH.mean()
ax1 = plt.subplot2grid((4,3),(2,2))
plt.pie(g.value_counts(),colors=("y","g"), autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('pH')

#For sulphates
h=dataset.sulphates<dataset.sulphates.mean()
ax1 = plt.subplot2grid((4,3),(3,0))
plt.pie(h.value_counts(), colors=("y","g"),autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('sulphates')

# For alcohol
i=dataset.alcohol<dataset.alcohol.mean()
ax1 = plt.subplot2grid((4,3),(3,1))
plt.pie(i.value_counts(),colors=("y","g"), autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('alcohol')

#For quantity
j=dataset.quality<dataset.quality.mean()
ax1 = plt.subplot2grid((4,3),(3,2))
plt.pie(j.value_counts(),colors=("y","g"),autopct = '%.2f%%',explode= explode,labels= lb)
plt.title('quality')
```

- From the pie charts, it can be observed that

- This followed by the chlorides (70.17%)

### Box plot of each column

```python
plt.figure(figsize=(10,15))
for i, col in enumerate(list(dataset.columns.values)):
    plt.subplot(4,3,i+1)
    dataset.boxplot(col)
    plt.grid()
    plt.tight_layout()
```

- The points outside the 1.5-IQR range or more above the third quartile are considered outliers.
- All other features have outliers besides alcohol.
